```python
print('조회수로 내림차순 정렬')
print(df_cat_med['view_count'].sort_values(ascending=False))
print()
print('좋아요 수로 내림차순 정렬')
print(df_cat_med['likes'].sort_values(ascending=False))
print() 
print('싫어요 수로 내림차순 정렬')
print(df_cat_med['dislikes'].sort_values(ascending=False))
print() 
print('댓글 수로 내림차순 정렬')
print(df_cat_med['comment_count'].sort_values(ascending=False))
print()
```

```python
# 조회수 / 좋아요
# total보다 위에 있으면 다른 분류들과 비교해 상대적으로 좋아요보다 조회수가 더 많음 (상대적으로 조회수에 비해 좋아요가 안 나온다고도 볼 수 있음.)
df_cat_med['view_div_likes'] = df_cat_med['view_count']/df_cat_med['likes']
df_cat_med['view_div_likes'].sort_values(ascending=False)

# 조회수 / 좋아요
# total보다 위에 있으면 다른 분류들과 비교해 상대적으로 좋아요보다 조회수가 더 많음 (상대적으로 조회수에 비해 좋아요가 안 나온다고도 볼 수 있음.)
#Autos & Vehicles Sports News & Politics Science & Technology Gaming Entertainment
#Total Comedy Howto & Style Travel & Events  Film & Animation Education People & Blogs Nonprofits & Activism Pets & Animals  Music
```

```python
# 조회수 / 싫어요
# total보다 위에 있으면 다른 분류들과 비교해 상대적으로 싫어요보다 조회수가 더 많음 (상대적으로 조회수에 비해 싫어요가 덜 나온다고도 볼 수 있음.)
df_cat_med['view_div_dislikes'] = df_cat_med['view_count']/df_cat_med['dislikes']
df_cat_med['view_div_dislikes'].sort_values(ascending=False)
```

```python
# 조회수 / 댓글수
# total보다 위에 있으면 다른 분류들과 비교해 상대적으로 댓글보다 조회수가 더 많음 (상대적으로 조회수에 비해 댓글이 덜 달린다고도 볼 수 있음.)
df_cat_med['view_div_comment'] = df_cat_med['view_count']/df_cat_med['comment_count']
df_cat_med['view_div_comment'].sort_values(ascending=False)
```

```python
#먼저 실행되었지만 편의상 뒤에 둠
# 좋아요 / 싫어요
# total보다 위에 있는 애들은 다른 분류들과 비교해 상대적으로 싫어요보다 좋아요가 더 많음

df_cat_med['likes_rate'] = df_cat_med['likes']/df_cat_med['dislikes']
df_cat_med['likes_rate'].sort_values(ascending=False)
```

```python
#먼저 실행되었지만 편의상 뒤에 둠
# 좋아요 / 댓글
# total보다 위에 있는 애들은 다른 분류들과 비교해 상대적으로 댓글보다 좋아요가 더 많음
# 정치부분은 좋아요보다 댓글이 훨씬 많다

df_cat_med['likes_div_comment'] = df_cat_med['likes']/df_cat_med['comment_count']
df_cat_med['likes_div_comment'].sort_values(ascending=False)
```

```python
# 싫어요 / 댓글
# total보다 위에 있는 애들은 다른 분류들과 비교해 상대적으로 댓글보다 싫어요가 더 많음
# 여행 카테고리는 싫어요보다 댓글이 훨씬 많다
# 교육 카테고리는 싫어요가 댓글보다 많다

df_cat_med['dislikes_div_comment'] = df_cat_med['dislikes']/df_cat_med['comment_count']
df_cat_med['dislikes_div_comment'].sort_values(ascending=False)
```
