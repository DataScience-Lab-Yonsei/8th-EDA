```python
print(">>>>>>>Information for youtube<<<<<<<<<<<<<")
#기본 정보
print("number of row      :",df.shape[0])
print("number of column   :",df.shape[1])
print("names of the column:",list(df.columns))
print("summary:")
print(df.info())
print("\n")

#결측값이 하나라도 있는 column 추출
for col in (df.columns):
    if df[col].isnull().any():
        print(col,"has null content")
        
```

```python
#불필요한 정보 제거

print(df['thumbnail_link']) #분석에 상관 없을 것으로 예상
df.drop('thumbnail_link',axis=1)
```

```python
titles=df['title'].unique()
print("Titles: ",titles)
print("\n# of unique titles: ",len(titles))

iDs=df['video_id'].unique()
print("\n\nIDs:",iDs)
print("\n# of unique IDs: ",len(iDs)) #ID 개수가 더 적음

SortByID=df.sort_values(by='video_id')
print(SortByID.tail(30)) #video_id가 같은데 title이 여러 개인 경우: ENG 영문 자막

SortByTitle=df.sort_values(by='title')
print(SortByTitle.tail(30))
```

```python
No_comm=df.loc[(df['comments_disabled'] == True), :]
print(No_comm[['comments_disabled','comment_count']])

Exception=df.loc[(df['comments_disabled'] == True)&df['comment_count']!=0, :]
if len(Exception) ==0:  
  print("‘comments_disabled’열이 True라면, ‘comment_count’도 0") 
else:
    print("‘comments_disabled’열이 True이지만, ‘comment_count가 0이 아닌 경우 존재")
```

```python
No_com=df.loc[(df['comment_count'] == 0), :]
print(No_com[['comments_disabled','comment_count']])

Exception=df.loc[(df['comment_count'] == 0)&df['comments_disabled']!=True, :]
if len(Exception) ==0:  
  print("‘comment_count’열이 0이라면, ‘comments_disabled’도 0") 
else:
    print("‘comment_count’열이 0이지만, ‘comments_disabled'가 0이 아닌 경우 존재")
```

```python
# 결측값 메꾸기
df.fillna(method='ffill')
df.to_csv('changed.csv', sep='\t', index=False)
```
