#### 광주

```python
# 웹 드라이버 객체 생성
browser = webdriver.Chrome('chromedriver', options = options)
```

```python
browser.get('https://sg.sbiz.or.kr/godo/stat/area.sg')
```

```python
#시군구 선택 
browser.find_element("xpath","/html/body/div[1]/div[2]/div[1]/div[2]/div/div/div[1]/div/div[1]/ul/li[1]/div/div[2]/a").click()
```

```python
#광주 선택
browser.find_element("xpath", "/html/body/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/ul/li[6]/label").click()
```

```python
#확인 선택
browser.find_element("xpath","/html/body/div[2]/div/div[2]/div/div[2]/a[2]/span").click()
```

```python
#현황보기 선택 
browser.find_element("xpath","/html/body/div[1]/div[2]/div[1]/div[2]/div/div/div[1]/div/div[2]/a").click()
```

```python
kj_table = browser.find_elements(By.CSS_SELECTOR,'#tableBody > tr > td')
kj_result = []
for a in kj_table:
  kj_result.append(a.text)
```

```python
#리스트 분할 함수
def list_chunk(lst, n):
  return [lst[i:i+n] for i in range(0, len(lst), n)]
```

```python
kj_result = list_chunk(kj_result, 9) # 한번만 실행할 것!
kj_result
```

```python
mycolumns = ["행정구역", "총가구수", "총인구수", "주요시설수", "집객시설수", "아파트 면적(1m²)당 기준시가(원)", "업소수", "업소당 가구수", "업소당 시설물수 "]
```

```python
kj_df = pd.DataFrame(kj_result, columns = mycolumns)
```

```python
kj_df #완료
```

```python
kj_df.to_csv("광주_지역현황.csv", index = None) #csv 파일로 내보내기
```
