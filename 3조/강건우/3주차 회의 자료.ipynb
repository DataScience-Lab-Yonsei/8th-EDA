```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Train=pd.read_csv('C:\\Users\\ADMIN\\Downloads\\data_ID.csv')
Train
```

```python
Train = Train[(Train['family_size'] <= 7)]
Train = Train.reset_index(drop=True)
Train
```

```python
def numeric_process(data):
  
   
    data['BIRTH_MINUS_EMPLOYED'] = data['DAYS_BIRTH'] - data['DAYS_EMPLOYED']
    
    return data 
```

```python
Train1 = numeric_process(Train)
Train1
```

```python
Train1.loc[Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.1), 'BME_percentile'] = 1

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.1)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.2)), 'BME_percentile'] = 2

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.2)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.3)), 'BME_percentile'] = 3

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.3)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.4)), 'BME_percentile'] = 4

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.4)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.5)), 'BME_percentile'] = 5

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.5)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.6)), 'BME_percentile'] = 6

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.6)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.7)), 'BME_percentile'] = 7

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.7)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.8)), 'BME_percentile'] = 8

Train1.loc[(Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.8)) & 
         (Train1['BIRTH_MINUS_EMPLOYED'] <= Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.9)), 'BME_percentile'] = 9

Train1.loc[Train1['BIRTH_MINUS_EMPLOYED'] > Train1['BIRTH_MINUS_EMPLOYED'].quantile(0.9), 'BME_percentile'] = 10

print(Train1['BME_percentile'].describe())
```

```python
Train1.groupby('BME_percentile')['credit_avg'].mean().plot(kind = 'bar')
plt.ylim()
plt.xlabel('BME_percentile')
plt.ylabel('average credit level')
plt.title('Average Credit Level VS BME_percentile')
```
